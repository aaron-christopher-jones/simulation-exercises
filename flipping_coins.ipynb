# Flipping So Many Coins

```python
import matplotlib.pyplot as plt
import numpy
import seaborn
```

```python
plt.rcParams["figure.figsize"] = (20,10)
```

```python
seaborn.set(font_scale=2.5)
seaborn.set_style("whitegrid")
```

```python
def coin_toss(n):
    """
    Calculates the cumulative proportion of heads
    given n flips of a fair coin.
    """
    flip = numpy.random.choice([0, 1], size=n, replace=True, p=[0.5, 0.5])
    average = [numpy.mean(flip[0:(f + 1)]) for f in range(len(flip))]
    return average
```

```python
num_simulations = 500
num_of_flips = 10000
```

```python
Trials = [coin_toss(num_of_flips) for _ in range(num_simulations)]
```

```python
for idx, vec in enumerate(Trials):
    
    plt.plot(Trials[idx], label="Trial{}".format(idx + 1), linewidth = 3)

plt.axhline(y=0.5, linestyle='dashed', linewidth=6, color='black')
plt.xlabel('Simulation Number')
plt.ylabel('Cumulative Proportion of Heads')
plt.show(block=True)
```
